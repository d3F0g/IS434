```python
#Import libraries
import matplotlib.pyplot as plt
from igraph import *
import igraph
import pandas as pd
import networkx as nx
import numpy as np
```

```python
G = nx.read_gml("../Week4/DATA/anna.gml")
#Find the clustering coefficient of the nodes
nx.clustering(G) # for undirect type
#Extract the coefficients out from the clustering dictionary
clustering = list(nx.clustering(G).values())

print("Anna_GML:")
print(nx.info(G))
```

WOW I LOVE SOCIAL ANALYTICS

```python
#Create a list that comtains the color choices for different degrees:
color_map = []
for i in range(len(clustering)):
    
    if clustering[i] >= 0.5:
        color_map.append('steelblue')
    
    elif clustering[i] >= 0.3 and clustering[i] < 0.5:
        color_map.append('skyblue')
    
    else:
        color_map.append('paleturquoise')

    
#Create size argument
size = []
for i in range(len(clustering)):
    size.append(clustering[i]*1000)  
```

```python
plt.figure(3)
#Set plot layout, k = 1 which means the nodes are somehow apart from each other 
pos = nx.circular_layout(G)


#Set nodes size based on degree
nx.draw_networkx_nodes(G, pos, node_size = size, node_color = color_map)



#Draw edges based on the two weight levels
nx.draw_networkx_edges(G, pos, width=1.5, alpha=0.5, edge_color = 'peachpuff')



#nx.draw(got, pos, with_labels = True)
nx.draw_networkx_labels(G, pos, font_size=8, font_color='k', font_family='sans-serif')


plt.axis('off') #Hide the axis
plt.show()
```

```python
print("Density: ", nx.density(G))
print("Diameter: ", nx.diameter(G))
print("Average path length: ", nx.average_shortest_path_length(G))
```

```python
degree_sq = list(G.degree())
avg_degree = np.mean(np.array(degree_sq)[:,1])
med_degree = np.median(np.array(degree_sq)[:,1])

max_degree = max(np.array(degree_sq)[:,1])
min_degree = np.min(np.array(degree_sq)[:,1])

print("Maximum degree : " + str(max_degree))
print("Minimum degree : " + str(min_degree))

print("Average degree : " + str(avg_degree))
print("Median degree : " + str(med_degree))
```

```python
degree_freq = np.array(nx.degree_histogram(G)).astype('float')

plt.figure(figsize=(12, 8))
plt.stem(degree_freq)
plt.ylabel("Frequence")
plt.xlabel("Degre")
plt.show()
```
